```python
df['SalesDate'] = pd.to_datetime(df['SalesDate'])
product_sales = df.groupby('Description')['SalesDollars'].sum().sort_values(ascending=False)
```

```python
sns.countplot(x='Classification',data=df)
plt.title('Classification')
plt.show()
```

```python
product_sales.head(10).plot(kind='barh', title='Top 10 Products by Sales')
plt.xlabel('Sales Dollars')
plt.ylabel('Product Description')
plt.tight_layout()
plt.show()
```

```python
corr = df[['SalesQuantity', 'SalesDollars', 'SalesPrice', 'ExciseTax', 'Volume']].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
Top 10 products by sales
top_products = df.groupby('Description')['SalesDollars'].sum().sort_values(ascending=False).head(10)

# Now plot
top_products.plot(kind='barh', figsize=(10,6), color='skyblue')
plt.title("Top 10 Products by Sales")
plt.xlabel("Sales Dollars")
plt.tight_layout()
plt.show()
```
